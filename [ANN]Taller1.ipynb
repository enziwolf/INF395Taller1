# Dataset 1 Sintético

Para el dataset sintético se elaboró una función continua con representación uni dimensional.

La función implementada corresponde a:

$$
f(x) = cos(x) + N(0,0.05) \,\, \forall x \in [-4\pi,4\pi]
$$

El intervalo es aproximado en los extremos porque se utiliza $\pi = 3.14$. Además se agregó un ruido normal para que el comportamiento de la función no fuera perfecto.

Como se mencionó con anterioridad se trabajará utilizando una red neuronal FeedForward y una red Convolucional para comparar el desempeño de estas.

Este dataset presenta una función sencilla con representación unidimensional y además cuenta con 1 solo regresor. 

Como este problema es sencillo una topología muy profunda puede caer fácilmente en el overfitting. Tras un poco de experimentación y búsqueda en internet, se determinó una topología adecuada para este problema corresponde a 5 capas ocultas, además de la capa de entrada y salida.

Para la capa de entrada no se tiene función de activación ya que solo recibe los datos.

Para las capas ocultas se utilizan 32 neuronas en cada capa (tuvo mejores resultados que de forma piramidal invertida) y, tras experimentación, se determinó que las mejores funciones de activación son *tanh* y *relu*. Los resultados obtenidos con ambas funciones son muy similares y no se observa una ventaja de una sobre la otra así que se optó por *tanh* como la función de activación en todas las capas ocultas. 

Para la capa de salida se tiene una activación lineal ya que es un problema de regresión.

```
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set()

#Datos originales
x = np.arange(start = -12.56, stop = 12.56, step = 0.01)

def funcion(x):
    return np.cos(x)

funcion_vec = np.vectorize(funcion)

#Creación de subconjuntos de datos
np.random.seed(10)
outside_range_x = np.arange(start = 12.56, stop = 18.84,step=0.06)
outside_range_y = funcion_vec(outside_range_x) + np.random.normal(loc=0,scale=0.05,size=outside_range_x.shape[0])


np.random.seed(100)
y = funcion_vec(x) + np.random.normal(loc=0,scale=0.05,size=x.shape[0])

plt.scatter(x=x,y=y)
plt.show()

x_train = []
y_train = []
x_test = []
y_test = []
x_val = []
y_val = []
for i in range(len(x)):
    if i%4 == 1:
        x_val.append(x[i])
        y_val.append(y[i])
    
    elif i%10 == 1:
        x_test.append(x[i])
        y_test.append(y[i])
    else:
        x_train.append(x[i])
        y_train.append(y[i])

x_train = np.array(x_train).reshape(len(x_train),1)
y_train = np.array(y_train).reshape(len(x_train),1)
x_test = np.array(x_test).reshape(len(x_test),1)
y_test = np.array(y_test).reshape(len(x_test),1)
x_val = np.array(x_val).reshape(len(x_val),1)
y_val = np.array(y_val).reshape(len(x_val),1)
outside_range_y = outside_range_y.reshape(outside_range_y.shape[0],1)

a = len(x_train)
b = len(x_val)
c = len(x_test)

print("Train:",a,"\nVal",b,"\nTest",c)
```

En el gráfico anterior se muestra todo el conjunto de datos con el que se va a trabajar. La idea de repetir varias veces el ciclo es ver si esto le da algo de ventaja a la red Convolucional ya que tiende a aprender patrones en los datos.

La división del dataset se hizo en 3 conjuntos:
* Train, para entrenar la red
* Validation, para ver el resultado del entrenamiento
* Test, para evaluación final

Además se creó un conjunto aparte fuera del rango de los datos de entrenamiento con el fin de evaluar si alguna de las dos arquitecturas tiende a mantener el patrón. Se espera que ninguna de las dos redes se desempeñe correctamente en este conjunto de datos, pero se evaluará que comportamiento es el "más malo".

```
plt.scatter(x=x_train,y=y_train,label="Train",color="black", s= 30)
plt.scatter(x=x_val,y=y_val,label="Validation",color="red",s=25)
plt.scatter(x=x_test,y=y_test,label="Test",color="blue",s=25)
plt.scatter(x=outside_range_x,y=outside_range_y,color="yellow",label="Outside")
plt.legend(loc="best")
plt.title("Division Conjuntos")
plt.show()
```

# FeedForward

La topología ya fue mencionada. Las capas en este caso son 5 capas ocultas densas de 32 unidades con función de activación *tanh*.

```
import numpy as np
from keras import models
from keras import layers
import time


# Crear la Red
network = models.Sequential()

#Input Layer
network.add(layers.Dense(units=32, input_shape=(1,)))

#Hidden Layers
network.add(layers.Dense(units=32,activation="tanh"))
network.add(layers.Dense(units=32,activation="tanh"))
network.add(layers.Dense(units=32,activation="tanh"))
network.add(layers.Dense(units=32,activation="tanh"))
network.add(layers.Dense(units=32,activation="tanh"))


#Output Layer, como es regresion es un output lineal (no hay activacion)
network.add(layers.Dense(units=1))

network.compile(loss='mse', 
                optimizer='adam', 
                metrics=['mse']) 

start = time.time()
history = network.fit(x_train,
                      y_train, 
                      epochs=60, 
                      verbose=0, 
                      batch_size=128,
                      validation_data=(x_val, y_val))

done = time.time()
print("Tiempo de entrenamiento: ",done-start,"[s]")

print(history.history.keys())

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()


### Post training
predictions = network.predict(x_test)
n = predictions.shape[0]
mse = sum((predictions-y_test)**2)/n

plt.scatter(x=x_test,y=predictions,label="Predictions")
plt.scatter(x=x_test,y=y_test,label="True")
plt.legend()
plt.title("FeedForward Mse: "+str(mse))
plt.show()


###Outside Set
predictions = network.predict(outside_range_x)
n = predictions.shape[0]
mse = sum((predictions-outside_range_y)**2)/n

plt.scatter(x=outside_range_x,y=predictions,label="Predictions")
plt.scatter(x=outside_range_x,y=outside_range_y,label="True")
plt.legend()
plt.title("FeedForward Mse Outside: "+str(mse))
plt.show()
```

### <u>Resultados y Comentarios</u>

Se aprecia un buen ajuste a los datos por parte de la red feedforward. La loss decae constantemente y, a simple vista, no se aprecia oscilación por lo que puede descartarse la presencia de overfitting. 

En el caso de prueba con datos nunca antes vistos y fuera del dominio original se obtuvo lo esperado. El modelo tiene un comportamiento malo y aleatorio en este conjunto. 

# CNN

La arquitectura convolucional cuenta con 5 capas ocultas, pero 4 de ellas son capas convolucionales de 32 filtros con activación *tanh* y la $5^{ta}$ capa oculta es una capa densa con la misma función de activación. 

La idea es que la principal diferencia entre ambas redes sea simplemente el aspecto convolucional, para descartar cualquier otro factor que pueda dar ventaja a una red sobre otra. 

```
from keras.models import Sequential
from keras.layers import Dense, Conv1D, Flatten
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import time

#Para la red convolucional hay que hacer un reshape a 3D de tipo: Nelementos * Natributos * Ncanales
#Como son filas se pasara Ncanales = 1 porque no hay otro.
x_train = x_train.reshape(x_train.shape[0],1,1)
y_train = y_train.reshape(y_train.shape[0],1,1)
x_test = x_test.reshape(x_test.shape[0],1,1)
y_test = y_test.reshape(y_test.shape[0],1,1)
x_val = x_val.reshape(x_val.shape[0],1,1)
y_val = y_val.reshape(y_val.shape[0],1,1)
outside_range_x = outside_range_x.reshape(outside_range_x.shape[0],1,1)
outside_range_y = outside_range_y.reshape(outside_range_y.shape[0],1,1)

model = Sequential()
#Input layer
model.add(Conv1D(32, 1, input_shape=(1,1)))
#Hidden Layers
model.add(Conv1D(32,1,activation="tanh"))
model.add(Conv1D(32,1,activation="tanh"))
model.add(Conv1D(32,1,activation="tanh"))
model.add(Conv1D(32,1,activation="tanh"))
model.add(Dense(32,activation="tanh"))
#Output Layer
model.add(Dense(1))
model.compile(loss="mse", optimizer="adam")

start = time.time()
history = model.fit(x_train, 
                      y_train,
                      epochs=60, 
                      verbose=0, 
                      batch_size=128,
                      validation_data=(x_val, y_val))

done = time.time()
print("Tiempo de entrenamiento: ",done-start,"[s]")

print(history.history.keys())

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()


### Post training
predictions = model.predict(x_test)
n = predictions.shape[0]
mse = sum((predictions-y_test)**2)/n

plt.scatter(x=x_test,y=predictions,label="Predictions")
plt.scatter(x=x_test,y=y_test,label="True")
plt.legend(loc="best")
plt.title("CNN Mse: "+str(mse))
plt.show()


###Outside Set
predictions = model.predict(outside_range_x)
n = predictions.shape[0]
mse = sum((predictions-outside_range_y)**2)/n

plt.scatter(x=outside_range_x,y=predictions,label="Predictions")
plt.scatter(x=outside_range_x,y=outside_range_y,label="True")
plt.legend(loc="best")
plt.title("CNN Mse Outside: "+str(mse))
plt.show()
```

### <u>Resultados y Comentarios</u> 

El comportamiento es bastante similar a la arquitectura feedforward, aunque se aprecia un mayor tiempo de entrenamiento y mayor MSE. El ajuste es bueno y tampoco se aprecia overfitting durante el entrenamiento. 

El comportamiento en el conjunto de datos "outside" es muy variable en ambas redes como para determinar si se aplica el patrón por casualidad o no. Tanto CNN como FF predicen muy mal el conjunto outside (como se esperaba).

A partir de lo anterior la red FF tiene la ventaja en este dataset simple.

Como comentario post-experimentación: el optimizador utilizado para las redes influyó drásticamente en el resultado. Originalmente la red FF estaba con RMSProp y la CNN con Adam. En ese caso la red FF presentaba overfitting y oscilación constante en la función de pérdida dándole la victoria a la CNN. Asi que destacamos la importancia del optimizador que utiliza la red.

```
import pandas as pd

filepath_dict = {'yelp':   'sentiment labelled sentences/yelp_labelled.txt'}

df_list = []
for source, filepath in filepath_dict.items():
    df = pd.read_csv(filepath, names=['sentence', 'label'], sep='\t')  
    df_list.append(df)

df = pd.concat(df_list)
df.head()
```

A continuación el dataset se separa en entrenamiento y validación:

```
from sklearn.model_selection import train_test_split

sentences = df['sentence'].values
y = df['label'].values

df_train_text, df_val_text, labels_train, labels_val = train_test_split(sentences, y, test_size=0.20, random_state=1000)
```

Luego, se preprocesan los textos. Lemmatizer reduce una palabra a sólo su lema, como por ejemplo "biblioteca" y "bibliografía" tienen lema "bibli". Lo anterior permite reducir el número de palabras a analizar al agruparlas en lemas. También se eliminan las stopwords, las cuales son palabras muy pequeñas que no aportan nada, como por ejemplo artículos, preposiciones, etc. También se reduce el texto a minúsculas y se eliminan los números para facilitar el procesamiento posterior.

```
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
import re, time
from nltk.corpus import stopwords
from nltk import WordNetLemmatizer, word_tokenize
def base_word(word):
    wordlemmatizer = WordNetLemmatizer()
    return wordlemmatizer.lemmatize(word) 
def word_extractor(text):
    commonwords = stopwords.words('english') #ya que los textos están en ingles
    text = re.sub(r'([a-z])\1+', r'\1\1',text)
    words = ""
    wordtokens = [ base_word(word.lower()) for word in word_tokenize(text) ]
    for word in wordtokens:
        if word not in commonwords: #se borran las stopwords
            words+=" "+word
    return words

texts_train = [word_extractor(text) for text in df_train_text]
texts_val = [word_extractor(text) for text in df_val_text]
```

A continuación se trabajará este problema en dos partes: Usando Term Frecuency que mantiene la estructura 1D; y Embedding que le agrega una dimensión más al establecer relaciones entre palabras (2D).

## Term Frecuency

La representación es una matriz de N (oraciones) x M (palabras) donde en la posicion n,m hay un k si la palabra está k veces en la oración o no.

Esto rompe la estructura del texto ya que no importa el orden en el que aparecen las palabras, sólo si se encuentran en la oración y con qué frecuencia.

```
#representación vectorial (term frequency)

import numpy as np
from sklearn.feature_extraction.text import CountVectorizer

#train
vectorizer = CountVectorizer(ngram_range=(1, 1), binary=False) #TF representation
vectorizer.fit(texts_train)
features_train = vectorizer.transform(texts_train)
vocab = vectorizer.get_feature_names()
dist=list(np.array(features_train.sum(axis=0)).reshape(-1,))

#validation
features_val = vectorizer.transform(texts_val)
vocab_v = vectorizer.get_feature_names()
distval=list(np.array(features_val.sum(axis=0)).reshape(-1,))


#10 palabras más frecuentes
used=features_train.sum(axis=0)
words_freq = [(word, used[0, idx]) for word, idx in     vectorizer.vocabulary_.items()]
words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
most=words_freq[:10]
print("Palabras más comunes:")
for word, freq in most:
    print(word, freq)

## Conjuntos de entrenamiento y validación
x_train= features_train.toarray()
y_train= labels_train.reshape(len(x_train),1)
x_val= features_val.toarray()
y_val= labels_val.reshape(len(x_val),1)

## Reshape de los conjuntos
x_train = x_train.reshape(x_train.shape[0],x_train.shape[1],1)
x_val = x_val.reshape(x_val.shape[0],x_val.shape[1],1)
```

```
from __future__ import print_function

from keras import layers
from keras.preprocessing import sequence
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.layers import Embedding
from keras.layers import Conv1D, GlobalMaxPooling1D,Flatten,MaxPooling1D
from keras.datasets import imdb
from keras.optimizers import SGD
import matplotlib.pyplot as plt
plt.style.use('ggplot')

def plot_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    x = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(x, acc, 'b', label='Training acc')
    plt.plot(x, val_acc, 'r', label='Validation acc')
    plt.title('Training and validation accuracy')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(x, loss, 'b', label='Training loss')
    plt.plot(x, val_loss, 'r', label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()
```

Las redes FF y CNN se diseñaran bajo un mismo "esqueleto": 1 capa densa o cov1d de entrada; 3 capas ocultas densas o conv1d dependiendo del caso, y con 4 filtros las primeras 2 y 8 filtros la tercera, todas con activación "relu"; y una capa de salida con 1 filtro con activación "sigmoid" el cual es el mejor para el caso de 2 categorias.

### Usando Red FF

```
batch_size = 64
epochs = 25
tipo1 = 'relu'
tipo2= 'tanh'
tipo3= 'sigmoid'

model = Sequential()

#Capa de Entrada
model.add(Dense(4, input_shape=(x_train.shape[1],1)))

#Capas Ocultas
model.add(Dense(4,  activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(4,  activation='relu'))
model.add(Dropout(0.5))

model.add(Flatten())
model.add(Dense(8))
model.add(Dropout(0.5))

#Capa de Salida
model.add(Dense(1))
model.add(Activation(tipo3))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

Por último, la capa de input de red neuronal realizará un Embedding el cual condesará las oraciones para realizar un proceso más rápido, dándole pesos a las palabras que estén relasionadas entre sí, es decir, no rompe las relaciones como lo hacía term frecuency. Podemos aprovechar lo anterior utilizando un modelo mucho más grande y preciso ya entrenado el cual servirá para pre-entrenar las oraciones durante el embedding, otorgándole pesos más precisos a la matriz de embedding.

```
import numpy as np

def create_embedding_matrix(filepath, word_index, embedding_dim):
    vocab_size = len(word_index) + 1  # Adding again 1 because of reserved 0 index
    embedding_matrix = np.zeros((vocab_size, embedding_dim))

    with open(filepath) as f:
        for line in f:
            word, *vector = line.split()
            if word in word_index:
                idx = word_index[word] 
                embedding_matrix[idx] = np.array(
                    vector, dtype=np.float32)[:embedding_dim]

    return embedding_matrix
embedding_dim = 50 ## dimensión del embedding
embedding_matrix = create_embedding_matrix(
    'glove.6B.50d.txt',
    tokenizer.word_index, embedding_dim)
nonzero_elements = np.count_nonzero(np.count_nonzero(embedding_matrix, axis=1))
nonzero_elements / vocab_size
```

De lo anterior se ve que el 97% del vocabulario obtenido con tokenizer es compatible con el vocabulario aportado por Glove.

### Usando Red FF

```
from __future__ import print_function

from keras.preprocessing import sequence
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.layers import Embedding
from keras.layers import Conv1D, GlobalMaxPooling1D
from keras.datasets import imdb
from keras import layers
import matplotlib.pyplot as plt
plt.style.use('ggplot')

def plot_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    x = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(x, acc, 'b', label='Training acc')
    plt.plot(x, val_acc, 'r', label='Validation acc')
    plt.title('Training and validation accuracy')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(x, loss, 'b', label='Training loss')
    plt.plot(x, val_loss, 'r', label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()

```

```
model = Sequential()
model.add(layers.Embedding(vocab_size, embedding_dim, 
                           weights=[embedding_matrix], 
                           input_length=maxlen, 
                           trainable=True))
model.add(layers.GlobalMaxPool1D())
##capas ocultas
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(8, activation='relu'))
model.add(layers.Dense(4, activation='relu'))
model.add(layers.Dense(2, activation='relu'))
##capa de salida
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.summary()
```

La red presentó un 71% de precisión en el conjunto de validación, pero claramente del gráfico loss vs epoch se puede apreciar una fuerte presencia de overfitting a partir del epoch 6, disparándose el loss de validación mientras que la loss de entrenamiento sólo decrece.

El overfitting se puede deber a distintos motivos: No hay suficientes datos de entrenamiento; el dataset puede que tenga inputs que no permitan generalizar correctamente; mala configuración de los hiperparámetros. Sobre los primeros dos problemas son descartables ya que el dataset consistió de 1000 oraciones, donde 800 se usaron para el entrenamiento; a la vez, el dataset estaba compuesto de un 50% con resultados positivos y la otra mitad con resultados negativos. Por ende sólo queda asumir que los hiperparámetros pueden ser mejorados.

Aún así, la presencia del overfitting no impide la comparación entre las redes feed forward y convolucional. De hecho, el overfitting es otra herramienta más para demostrar que una red es mejor que otra, siempre y cuando mantengamos una misma arquitectura para ambas redes.
```
def getpos(array):
    #Del OH-Encoding obtiene el tag (en numero)
    pos = 0
    for elem in array:
        if elem == 1:
            return pos
        pos +=1
    return -1

def get_tags(y):
    #Dado una lista de OHEs retorna una lista de sus tags
    tags = list()
    for array in y:
        pos = getpos(array)
        tags.append(pos)
    return tags

def count_tags(y):
    total = {}
    for array in y:
        tag = getpos(array)
        if tag not in total.keys():
            total[tag] = 1
        else:
            total[tag] +=1
    return total

count_tags(new_oh_output)
```

Un resúmen del trabajo que se hace es generar una máscara con los índices de los datos que serán del conjunto de prueba (el resto de entrenamiento). Para cada label (0 a 9) se hace un muestreo del p%, lo que, en teoría, se transforma en un p% de datos para el dataset de prueba, pero manteniendo, la distribución de las clases en cada dataset. 

El dataset original no viene ordenado, asi que ambos datasets, prueba y entrenamiento, estarán "aleatorizados" en orden. 


```
# Split del dataset, a mano
def get_indexes(tag,tag_list):
    #retorna una lista de los indices de tag_list que coinciden con tag
    indexes = list()
    for t in range(len(tag_list)):
        if tag_list[t] == tag:
            indexes.append(t)
    
    return indexes

def array_of_indexes(y):
    #Genera un diccionario con los indices de cada tag de los OHE's
    tag_indexes = {}
    tag_list = get_tags(y)
    unique_tags = set(tag_list)
    for ut in unique_tags:
        tag_indexes[ut] = get_indexes(ut,tag_list)
    return tag_indexes

diccio = array_of_indexes(new_oh_output)

def generate_mask(arr_of_i,p):
    test_index = list()
    for _,v in arr_of_i.items():
        sample_size = int(len(v)*p)
        sample = np.random.choice(v,size=sample_size,replace=False)
        test_index.extend(sample)
    return test_index

t_indexes = sorted(generate_mask(diccio,0.15)) #15% para test (mejor score logrado con 15% (es estocastico si))

def split_dataset(mask,x,y):
    #X e Y son de misma cantidad de filas, mask son los indices que se calcularon para la prueba
    train_x = list()
    train_y = list()
    test_x = list()
    test_y = list()
    for index in range(len(x)):
        if (len(mask)> 0) and (index == mask[0]):
            mask.pop(0)
            test_x.append(x[index])
            test_y.append(y[index])
        else:
            train_x.append(x[index])
            train_y.append(y[index])

    train_x = np.stack(train_x)
    train_y = np.stack(train_y)
    test_x = np.stack(test_x)
    test_y = np.stack(test_y)

    return train_x,train_y,test_x,test_y 

x_train,y_train,x_test,y_test = split_dataset(t_indexes,x_final,new_oh_output)

print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
#Aqui se puede llamar la function count_tags(y) en los splits.
R_mean = np.mean(x_train[:,:,:,0])
G_mean = np.mean(x_train[:,:,:,1])
B_mean = np.mean(x_train[:,:,:,2])
R_var = np.var(x_train[:,:,:,0])
G_var = np.var(x_train[:,:,:,1])
B_var = np.var(x_train[:,:,:,2])
```

### ¿Qué es transfer learning? (TL)

TL es la utilización de un modelo que fue entrenado en otro conjunto de datos para atacar un problema similar al que se busca resolver. En este caso, clasificación de imágenes. Con TL se puede crear un modelo base utilizando el modelo entrenado con pesos ajustado en un dataset de mayor magnitud que el que se tiene. Y luego agregar capaz que se adecuén al problema en particular.

Una imagen vale mas que 1000 palabras:

### Implementación del TL

Como se mencionó, se utilizará InceptionResNetV2 como el modelo base y se harán los siguientes cambios/agregaciones:

* El input será de 128x128x3 en vez de las dimensiones de imagenet 
* La capa de salida serán 10 neuronas con activación softmax, porque se tienen 10 clases y no 1000.
* Se crea una capa de normalización que lleva a los pixeles de un rango de (0,255) a uno mas acotado esto facilita su trabajo para la red.
* Se hace un data augmentation que aumenta la cantidad de los datos de entrenamiento al hacer pequeñas variaciones de este.

```
## Transfer learning, la idea es cargar un modelo, congelar las capas no originales, entrenar las nuevas y luego entrenar le modelo entero
from tensorflow import keras
from tensorflow.keras import layers

#Como tenemos pocos ejemplos es bueno hacer data augmentation que son variaciones de los datos de entrada
#aumentan el tamaño de prueba.
data_augmentation = keras.Sequential(
    [
        layers.experimental.preprocessing.RandomFlip("horizontal"),
        layers.experimental.preprocessing.RandomRotation(0.1),
    ]
)

base_model = keras.applications.InceptionResNetV2(weights="imagenet",input_shape=(128,128,3),include_top=False)
base_model.trainable = False

inputs = keras.Input(shape=(128, 128, 3))
x = data_augmentation(inputs) 

#Se agrega una capa de normalizacion
#Esto hace que los cálculos se ejecuten con mayor rapidez 
#El gradiente no se aloque por valores muy altos de pixeles
#La data queda centrada y las "features" quedan en un rango mas acotado
#Además, como hay compartición de pesos, esto ayuda a que la compartición sea mas facil
#porque asi se tienen dos pesos similares y no uno muy grande y otro muy chico.
norm_layer = keras.layers.experimental.preprocessing.Normalization()
mean = np.array([R_mean,G_mean,B_mean]) #se normaliza cada canal
var = np.array([R_var,G_var,B_var]) 
# Scale inputs 
x = norm_layer(x)
norm_layer.set_weights([mean, var]) #pesos iniciales son la media y varianza de cada canal


# The base model contains batchnorm layers. We want to keep them in inference mode
# when we unfreeze the base model for fine-tuning, so we make sure that the
# base_model is running in inference mode here.
x = base_model(x, training=False)
x = keras.layers.GlobalAveragePooling2D()(x)
x = keras.layers.Dropout(0.2)(x)  # Regularize with dropout, evita overfitting seteando los pesos de algunas a 0
#en cada step durante el entrenamiento
outputs = keras.layers.Dense(10, activation='softmax')(x) #capa de salida, 10 clases, como es excluyente se usa softmax.
model = keras.Model(inputs, outputs)

model.summary()

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])


model.fit(x_train, y_train,
          batch_size=32,
          epochs=20,
          verbose=1,
          validation_data=(x_test, y_test))


base_model.trainable = True
model.summary()

model.compile(
    optimizer=keras.optimizers.Adam(1e-5),  # Low learning rate
    loss=keras.losses.categorical_crossentropy,
     metrics=['accuracy']
)


history = model.fit(x_final, new_oh_output, #al saber que trabaja bien se entrena con todo el dataset
          batch_size=32,
          epochs=20,
          verbose=1)
```

```
#Accuracy en test
def max_pos(array):
    maximo = array[0]
    max_pos = 0
    for pos in range(1,len(array)):
        if array[pos] > maximo:
            maximo = array[pos]
            max_pos = pos
    return max_pos

def accuracy(preds,true):
    score = 0
    for p,t in zip(preds,true):
        if max_pos(p) == max_pos(t):
            score += 1
    return score/len(preds)

accuracy(model.predict(x_test),y_test)
#0.9607
```
